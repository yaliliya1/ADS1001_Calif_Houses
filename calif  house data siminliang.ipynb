```python
import pandas as pd
housing = pd.read_csv('california-housing-data.csv')
housing
```

```python
housing = housing.dropna(subset=['total_bedrooms'])
housing
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(data=housing)
housing.corr()   #get correlation coefficients


#
```

```python
plt.scatter(x=housing['longitude'],y=housing['latitude'])

#we can't see claerly
```

```python
plt.scatter(x=housing['longitude'],y=housing['latitude'],alpha=0.1)

#we konw the density of huoses from the visualization of geography 
```

```python
plt.scatter(x=housing['longitude'],y=housing['latitude'],s=housing['population']/80,label='population',alpha=0.4,c=housing['median_house_value'],cmap=plt.get_cmap('viridis'))
plt.colorbar()
plt.legend()
plt.show()

#The size (s) and color (c) of the points are determined by the population quantity (housing['population']) 
# and the median house value (housing['median_house_value']), respectively.
```

```python
value_income = plt.scatter(data=housing, x='median_income', y='median_house_value',alpha=0.07)
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```
